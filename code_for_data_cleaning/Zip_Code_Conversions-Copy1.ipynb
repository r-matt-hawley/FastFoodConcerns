```python
# import library
import pandas as pd
# uszipcode gives the state&county for each zipcode
from uszipcode import SearchEngine

# we are using the zipcode in "cleaned_population.csv" as the zipcode souce 
# the zipcode inside "cleaned_population.csv" is complete - this csv includes all the zipcode in USA
population_file = "cleaned_population.csv"
pop_df1 = pd.read_csv(population_file)

# Convert Zip Code from numpy.int64 to str
pop_df["zipcode_str"] = pop_df1['zip_code'].apply(lambda x: str(x).zfill(5))

#show the dataframe
pop_df.head()
```

```python
# create the blank list of state and county to store the data
state = []
county = []

# use counter to the process of the coding process
counter = 0

# Find county and state by zip code
#test_code = pop_df["zipcode_str"]
search = SearchEngine(simple_zipcode=True) # set simple_zipcode=False to use rich info database
for i in pop_df["zipcode_str"]:
    zipcode_data = search.by_zipcode(i)
    
    # convert the data into dictionery
    zipcode_dic = zipcode_data.to_dict()
    
    # store the "state" and "county" from zip dictionary into the lists
    state.append(zipcode_dic["state"])
    county.append(zipcode_dic["county"])
    
    # use the counter to show the storing progress
    counter += 1
    if ((counter % 100) == 0):
        print(f"completed {counter} out of {len(pop_df['zipcode_str'])}")

#store the data into pop_df dataframe
pop_df["state"] = state
pop_df["county"] = county
pop_df.head()
```

```python
# store the data in .csv file
pop_df.to_csv("cleaned_zip_population.csv")
```
